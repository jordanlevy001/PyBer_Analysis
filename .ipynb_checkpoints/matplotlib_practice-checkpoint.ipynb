```python
# Import dependencies.

import matplotlib
matplotlib.__version__


import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Set the x- axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars acumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with the MATLAB method
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with the object-oriented method: ax.plt().
# Add a legend for the city of Boston
# Change the line width to 2 points
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, linewidth=2, label='Boston')

ax.legend()

# Change line color to green

# Change the marker to a diamond

# Add a title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

# Add a y-axis limit
ax.set_ylim(0, 45)
# Add grid lines
ax.grid()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()


```

```python
# Bar Charts
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color='green', label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis, color="magenta", label='Boston')

# Create the plot
plt.gca().invert_yaxis()

plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Create a vertical bar chart with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')
```

```python
# Create a horizontal bar chart using the object-oriented interface method
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
ax.invert_yaxis()

ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title("PyBer Fare by Month")
ax.legend()
```
